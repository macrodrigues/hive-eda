# Exploratory Data Analysis on HIVE blockchain

```python
import pandas as pd
import seaborn as sns
import plotly as px
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# read the csv generated previously with beem HIVE package
df = pd.read_csv('hive_data.csv')
```

```python
#drop unnamed column
df.drop([df.columns[0]], axis = 1, inplace = True)
```

```python
#sort by 'pending value payout'
df = df.sort_values(by=df.columns[-2], ascending=False)
```

```python
df
```

## Categories that generate the most value per post

```python
df_cat = df.groupby(by=['Category']).sum().sort_values(by=df.columns[-2], ascending=False)
```

```python
df_cat.head(20)
```

```python
df_cat_best = df_cat.head(20)
```

```python
categories = [
    'Leo Finance',
    'Pinmapple',
    'Photography Lovers',
    'Foodies Bee Hive',
    'HIVE CN 中文社区',
    'Hive Gaming',
    "Hive Open Mic",
    "Full Deportes",
    'Liketu',
    'Music',
    'Alien Art Hive',
    'Wednesday Walk',
    'MakeUp Power',
    'Splinterlands',
    'Movies & TV Shows',
    'EspaVlog',
    'Ladies of Hive',
    'Geek Zone',
    'WEEKEND EXPERIENCES',
    'GEMS']
```

```python
df_cat_best.index
```

```python
rename_dict = dict(zip(list(df_cat_best.index), categories))

df_cat_best = df_cat_best.rename(index = rename_dict)
df_cat_best = df_cat_best.reset_index()
```

```python
df_cat_best
```

```python
plt.figure(figsize=(15,7))
ax = sns.barplot(x='Category', y='Pending payout value', data=df_cat_best)
for item in ax.get_xticklabels():
    item.set_rotation(60)
```

Above we can see the categories that generate the most rewards per post. This doesn't mean that these categories are the ones who generate the most value. It can even mean that these categories are highly unbalanced, rewarding some users much more than others.

## Correlations

```python
import ast

df_corr = df[['Reputation', 'Votes', 'Body Length', 'Tags', 'Pending payout value']]
df_corr['Tags'] = df_corr['Tags'].apply(lambda x: ast.literal_eval(x))
df_corr['Number of Tags'] = df_corr['Tags'].apply(lambda x: len(x))
```

```python
sns.heatmap(df_corr.corr(), cmap="YlGnBu");
```

From the heatmap above we can see that there aren't any strong correlation between features. The number of votes it's slightly correlated with the payout value. But reputation, number of tags and body length do not seem to affect the amount of rewards.

## What about the best tags?

```python
df_temp_tags = df_corr.explode('Tags').drop(['Length Tags', 'Reputation', 'Body Length'], axis=1)
df_temp_tags['Tag Counts'] = df_temp_tags['Tags'].apply(lambda x: 1)
```

```python
df_tags = df_temp_tags.groupby('Tags')\
    .sum().sort_values(by='Pending payout value', ascending=False)
df_tags
```

```python
sns.heatmap(df_tags.corr(), cmap="YlGnBu");
```

The most used tags, seem to influence the payout value, let's see which ones are the best!

```python
plt.figure(figsize=(10,7))
ax = sns.barplot(x='Tags', y='Pending payout value', data=df_tags.reset_index().head(10))
for item in ax.get_xticklabels():
    item.set_rotation(60)
```
